🔍 Hypothesis Testing
Hypothesis testing is a statistical method used to make decisions or inferences about population parameters based on sample data. It helps you determine whether a result is statistically significant.

🧪 Null Hypothesis (H₀) vs. Alternative Hypothesis (H₁)
Null Hypothesis (H₀): Assumes no effect or no difference; it's the default or status quo.

Alternative Hypothesis (H₁): Contradicts H₀; it suggests an effect or difference exists.

🎯 Significance Level (α)
Typically set at 0.05 (5%).

It's the threshold for rejecting H₀.

If the probability (P-value) of observing the data is less than α, you reject H₀.

📊 P-value
Measures the probability of observing the test results under the null hypothesis.

A small P-value (typically ≤ α) indicates strong evidence against H₀.

📉 Interpreting P-value
P ≤ α: Reject H₀ (evidence supports H₁).

P > α: Fail to reject H₀ (insufficient evidence against it).

❌ Type I and Type II Errors
Type I Error (α): Incorrectly rejecting H₀ when it’s true (false positive).

Type II Error (β): Failing to reject H₀ when it’s false (false negative).

🔁 One-tailed vs. Two-tailed Tests
One-tailed: Tests for an effect in one direction (e.g., greater than).

Two-tailed: Tests for any difference (either greater or less than).

🧮 Z-test
Used when population variance is known and sample size is large (n > 30).

Assumes data follows a normal distribution.

📈 T-distribution
Used when population variance is unknown and sample size is small (n < 30).

Wider tails than the normal distribution—accounts for more uncertainty.

⚖️ Z-test vs. T-test
Z-test: Known population variance and large sample.

T-test: Unknown variance and/or small sample.

🧪 T-test
Tests if there is a significant difference between group means.

Types: one-sample, two-sample (independent), and paired sample.

🔄 Z-test and T-test Relationship
Both test means; T-test becomes similar to Z-test as sample size increases.

📏 Confidence Interval
A range of values likely to contain the population parameter.

Common: 95% CI means there's a 95% chance the interval contains the true mean.

📐 Margin of Error
Reflects the range of uncertainty in the CI.

Affects the width of the interval; larger margin = less precision.

🔲 Chi-square Distribution
Used in tests for categorical data.

Skewed distribution that becomes more normal with higher degrees of freedom.

✅ Chi-square Goodness of Fit Test
Tests whether observed frequencies match expected frequencies.

Useful for determining if a sample fits a distribution.

⚖️ F-distribution
Used to compare variances between groups.

Right-skewed; used in ANOVA and F-tests.

📊 ANOVA (Analysis of Variance)
Tests for differences in means across 3+ groups.

Assumptions: independence, normality, equal variances.

🔄 Types of ANOVA
One-way ANOVA: One independent variable.

Two-way ANOVA: Two independent variables.

Repeated Measures ANOVA: Same subjects measured multiple times.

✅ F-test
Compares two variances or models (e.g., in ANOVA).

Determines whether a group of variables is collectively significant.
